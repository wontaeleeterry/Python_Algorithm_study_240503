```python
# 데이터프레임 복사

data_Highway_copy = data_Highway.copy()
```

```python
data_Highway_copy
```

```python
data_Highway_copy.speed[369700:369776]
```

```python
data_Highway_final = data_Highway_copy.loc[0:369775, 'time': 'speed']
data_Highway_final
```

```python
# 여러개의 결측치를 선형보간법으로 채우기

data_Highway_final['speed'] = data_Highway_final['speed'].interpolate(method = 'polynomial', order=1)

# order를 1로 변경하는 경우(241124)
```

```python
data_Highway_final
```

```python
# 소수점 둘째자리까지 정리 (241122)

column = 'speed'
where = 2

data_Highway_final[column]= round(data_Highway_final[column], where)
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(15,5))

plt.yticks(np.arange(0, 4001, 200))
plt.xticks(np.arange(0, 470001, 30000))

plt.plot(data_Highway_final.time, data_Highway_final.speed)

plt.title('City Mode')
plt.xlabel('Time (msec)')
plt.ylabel('TC Speed (Hz)')
plt.xlim([0, 380000])      # X축의 범위: [xmin, xmax]
plt.ylim([0, 4000])     # Y축의 범위: [ymin, ymax]
plt.grid(True, linestyle='--')
plt.show()
```
