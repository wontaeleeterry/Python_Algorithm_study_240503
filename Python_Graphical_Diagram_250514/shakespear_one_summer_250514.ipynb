```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.Graph()

# Add characters as nodes
characters = [
    "Theseus", "Hippolyta", "Egeus", "Hermia", "Lysander", "Demetrius", "Helena",
    "Oberon", "Titania", "Puck", "Nick Bottom", "Peter Quince", "Francis Flute",
    "Tom Snout", "Snug", "Robin Starveling"
]

G.add_nodes_from(characters)

# Add edges with relationship labels
relationships = [
    ("Theseus", "Hippolyta", "engaged"),
    ("Egeus", "Hermia", "father-daughter"),
    ("Hermia", "Lysander", "lovers"),
    ("Hermia", "Demetrius", "unwanted suitor"),
    ("Demetrius", "Helena", "eventual lovers"),
    ("Helena", "Hermia", "friends"),
    ("Lysander", "Demetrius", "rivals in love"),
    ("Oberon", "Titania", "married"),
    ("Oberon", "Puck", "servant"),
    ("Titania", "Nick Bottom", "enchanted lover"),
    ("Peter Quince", "Nick Bottom", "co-worker"),
    ("Peter Quince", "Francis Flute", "co-worker"),
    ("Peter Quince", "Tom Snout", "co-worker"),
    ("Peter Quince", "Snug", "co-worker"),
    ("Peter Quince", "Robin Starveling", "co-worker"),
    ("Nick Bottom", "Francis Flute", "co-worker"),
    ("Nick Bottom", "Tom Snout", "co-worker"),
    ("Nick Bottom", "Snug", "co-worker"),
    ("Nick Bottom", "Robin Starveling", "co-worker"),
]

# Add edges to the graph
for char1, char2, relation in relationships:
    G.add_edge(char1, char2, label=relation)

# Define relationship color mapping
relationship_colors = {
    "engaged": "orange",
    "father-daughter": "blue",
    "lovers": "red",
    "unwanted suitor": "purple",
    "eventual lovers": "green",
    "friends": "cyan",
    "rivals in love": "magenta",
    "married": "darkred",
    "servant": "brown",
    "enchanted lover": "pink",
    "co-worker": "gray"
}

# Extract edge colors based on the relationship type
edge_colors = [relationship_colors[d['label']] for u, v, d in G.edges(data=True)]

# Positioning the nodes using spring layout
pos = nx.spring_layout(G, seed=42)

# Draw the graph
plt.figure(figsize=(18, 12))
nx.draw(
    G,
    pos,
    with_labels=True,
    node_size=4000,          # Larger nodes
    node_color="lightblue",
    font_size=12,            # Larger font for node labels
    font_weight="bold",
    edge_color=edge_colors,
    width=2
)

# Draw edge labels
edge_labels = {(u, v): d['label'] for u, v, d in G.edges(data=True)}
nx.draw_networkx_edge_labels(
    G,
    pos,
    edge_labels=edge_labels,
    font_color="black",
    font_size=9
)

# Title and formatting
plt.title("Character Relationship Network in 'A Midsummer Night's Dream'\n(Edge Colors Indicate Relationship Types)", fontsize=16)
plt.axis("off")
plt.tight_layout()
plt.show()
```
