```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Generate a simple dataset: 100 random numbers from a normal distribution
data = np.random.normal(loc=10, scale=2, size=100)

# Bootstrap parameters
n_iterations = 1000  # Number of bootstrap samples
n_size = len(data)   # Size of each bootstrap sample

# Function to compute the statistic of interest (mean in this case)
def bootstrap_sample_mean(data, n_iterations):
    boot_means = np.empty(n_iterations)

    # Resample with replacement and compute means
    for i in range(n_iterations):
        sample = np.random.choice(data, size=n_size, replace=True)
        boot_means[i] = np.mean(sample)
    
    return boot_means

# Get bootstrap sample means
bootstrap_means = bootstrap_sample_mean(data, n_iterations)

# Compute original mean and confidence intervals (95%)
original_mean = np.mean(data)
confidence_interval = np.percentile(bootstrap_means, [2.5, 97.5])

# Plot the bootstrap distribution
plt.hist(bootstrap_means, bins=30, color='blue', alpha=0.7, label='Bootstrap Means')
plt.axvline(original_mean, color='red', linestyle='dashed', linewidth=2, label='Original Mean')
plt.axvline(confidence_interval[0], color='green', linestyle='dashed', linewidth=2, label='95% CI Lower')
plt.axvline(confidence_interval[1], color='green', linestyle='dashed', linewidth=2, label='95% CI Upper')
plt.title('Bootstrap Distribution of Sample Means')
plt.xlabel('Mean')
plt.ylabel('Frequency')
plt.legend()
plt.show()

original_mean, confidence_interval
```
