```python
# 등락률 계산하기

import pymysql

con = pymysql.connect(
    user='root',
    passwd=PASS,  # 윈도우 노트북 확인
    host='127.0.0.1',
    db='db_sqlstk',        # 윈도우 노트북 확인
    charset='utf8'
)


mycursor = con.cursor()

query = '''
SELECT  T1.STK_CD ,T1.STK_NM
        ,T2.DT ,T2.C_PRC
        ,T3.DT ,T3.C_PRC
        ,ROUND((T2.C_PRC - T3.C_PRC) / T3.C_PRC * 100,2) CHG_RT   # 계산 결과가 CHG_RT 컬럼 값이 된다.
FROM    STOCK T1
        INNER JOIN HISTORY_DT T2
          ON (T2.STK_CD = T1.STK_CD)
        INNER JOIN HISTORY_DT T3
          ON (T3.STK_CD = T2.STK_CD)
#WHERE   T1.STK_NM LIKE '삼성%'                     # 삼성 계열사
WHERE   T1.STK_NM IN ('삼성전자','카카오','LG화학')     # 특정 종목 여러개 지정
AND     T2.DT = STR_TO_DATE('20201230','%Y%m%d')
AND     T3.DT = STR_TO_DATE('20200102','%Y%m%d');
'''
mycursor.execute(query)
data = mycursor.fetchall()
con.close()

# 쿼리 결과를 데이터프레임으로 출력
import pandas as pd

df = pd.DataFrame(data) 
df.columns = ['종목코드', '종목명', '일자(시작)', '종가', '일자(종료)', '종가', '수익률(%)']
df

#display(data)
```

```python
# 하나의 SQL에 인라인 뷰를 여러 개 사용할 수 있는 것처럼 WITH절도 여러 개 사용 가능
# RES1에서는 2019년 2월에 등락률이 가장 큰 종목 열 개를 찾아내고, 
# RES2에서는 RES1에서 찾은 결과 중에 2월 28일 가격이 가장 작은 종목 세 개를 찾아내고 있다.

import pymysql

con = pymysql.connect(
    user='root',
    passwd=PASS,
    host='127.0.0.1',
    db='db_sqlstk',
    charset='utf8'
)

mycursor = con.cursor()

query = '''
WITH RES1 AS (
    SELECT  T1.STK_CD ,T1.STK_NM
            ,T2.DT ,T2.C_PRC 0228_가격
            ,T3.C_PRC 0201_가격       
            ,ROUND((T2.C_PRC - T3.C_PRC) / T3.C_PRC * 100,2) CHG_RT
    FROM    STOCK T1
            INNER JOIN HISTORY_DT T2 ON (T2.STK_CD = T1.STK_CD)
            INNER JOIN HISTORY_DT T3 ON (T3.STK_CD = T1.STK_CD)
    WHERE   T2.DT = STR_TO_DATE('20190228','%Y%m%d')
    AND     T3.DT = STR_TO_DATE('20190201','%Y%m%d')
    ORDER BY CHG_RT DESC
    LIMIT 10)
,RES2 AS(
    SELECT  T4.*
    FROM    RES1 T4
    ORDER BY T4.0228_가격 ASC
    LIMIT 3)
SELECT  T5.*
FROM    RES2 T5;
'''
mycursor.execute(query)
data = mycursor.fetchall()
con.close()

# 쿼리 결과를 데이터프레임으로 출력
import pandas as pd

df = pd.DataFrame(data) 
df.columns = ['종목코드', '종목명', '일자', '종가(2/28)', '종가(2/01)', '수익률(%)']
df

#display(data)

# 가격이 가장 작은 종목 3개를 찾아 2월 가격 등락 확인
```

```python
# 2019년 3월에 거래 금액이 가장 큰 종목 Top 3 조회

import pymysql

con = pymysql.connect(
    user='root',
    passwd=PASS,
    host='127.0.0.1',
    db='db_sqlstk',
    charset='utf8'
)

mycursor = con.cursor()

query = '''
SELECT  T1.STK_CD ,MAX(T1.STK_NM) STK_NM
        ,ROUND(SUM(T2.C_PRC * T2.VOL)/1e8,1) VOL_AMT   # 거래금액 (1억 단위)로 계산 + 소수점 1자리 : 거래량 x 종가

FROM    STOCK T1
        INNER JOIN HISTORY_DT T2
           ON (T2.STK_CD = T1.STK_CD)
WHERE   T1.EX_CD = 'KD'                             # 코스닥 종목(KD)에서 선택

AND     T2.DT >= STR_TO_DATE('20190301','%Y%m%d')   # HISTORY 테이블의 날짜 조건 : 이 날짜 이후
AND     T2.DT < STR_TO_DATE('20190401','%Y%m%d')    # HISTORY 테이블의 날짜 조건 : 이 날짜 이전

GROUP BY T1.STK_CD   # ???? (20514)    # https://wikidocs.net/139147  GROUP BY 이해 필 (240514)
ORDER BY VOL_AMT DESC
LIMIT 3;      # 표시할 항목의 갯수
'''
mycursor.execute(query)
data = mycursor.fetchall()
con.close()

# 쿼리 결과를 데이터프레임으로 출력
import pandas as pd

df = pd.DataFrame(data) 
df.columns = ['STK_CD', 'STK_NM', 'VOL_AMT']
df
```

### INLINE VIEW (240517)
- 인라인 뷰(INLINE VIEW)는 단독으로 실행 가능한 SQL을 FROM절에 괄호로 묶어 사용하는 것
- SQL 안에 또 다른 SQL이 있는 것이기 때문에 서브쿼리의 한 종류로 FROM절에 사용된 서브쿼리 정도로 생각한다.
- 아래 SQL은 FROM절에 VW_STOCK_HISTORY 대신에 해당 뷰를 생성할 때 사용한 SQL을 괄호로 묶어 그대로 사용하고 있다.
- 출처 : https://wikidocs.net/147055

```python
# 2019년 3월에 거래금액이 가장 큰 종목 Top 3를 조회
# 이어서, 위 SQL의 결과에 2019년 4월 1일의 종가와 2019년 12월 30일 종가를 추가해서 보여주는 쿼리
PASS = "$$"

QUERY = '''
#[SQL 5.12.2-2]

SELECT  T1.STK_CD ,MAX(T1.STK_NM) STK_NM
        ,ROUND(SUM(T2.C_PRC * T2.VOL)/1e8,1) VOL_AMT    # 종가 X 거래량, 1억 단위로 변경 -> VOL_AMT 로 별칭 지정
FROM    STOCK T1
        INNER JOIN HISTORY_DT T2
           ON (T2.STK_CD = T1.STK_CD)         # 별칭 테이블 T1과 T2의 종목 코드가 일치하는 종목 
WHERE   T1.EX_CD = 'KD'                       # 조건문 : 코스닥(KD) 종목만 추출
AND     T2.DT >= STR_TO_DATE('20190301','%Y%m%d')    # 조건 추가(AND) : 특정 일자 이후
AND     T2.DT < STR_TO_DATE('20190401','%Y%m%d')     # 조건 추가(AND) : 틀정 일자 이전   WHERE + AND + AND
GROUP BY T1.STK_CD                                   # 그룹으로 묶는 습관
ORDER BY VOL_AMT DESC                                # 내림 차순으로 정리
LIMIT 10;                                             # 표시할 데이터의 갯수 지정
'''


# INLINE VIEW 실습
QUERY = '''
#[SQL 5.12.2-3]

# 아래 내부 쿼리로 생성한 테이블 T3에서 항목을 불러오고, 다른 테이블의 데이터를 추가

SELECT  T3.* ,T4.C_PRC C_PRC_0401 ,T5.C_PRC C_PRC_1230       # 내부에 정의된 T3 테이블의 모든 내용 가져오기 T3.*

# 내부에 쿼리가(완벽한) 하나 더 존재 -> 별칭을 부여한 테이블로 생성
FROM    (  
        SELECT  T1.STK_CD ,MAX(T1.STK_NM) STK_NM             #################### 내부 쿼리의 시작
                ,ROUND(SUM(T2.C_PRC * T2.VOL)/1e8,1) VOL_AMT
        FROM    STOCK T1
                INNER JOIN HISTORY_DT T2
                   ON (T2.STK_CD = T1.STK_CD)
        WHERE   T1.EX_CD = 'KD'
        AND     T2.DT >= STR_TO_DATE('20190501','%Y%m%d')
        AND     T2.DT < STR_TO_DATE('20190601','%Y%m%d')
        GROUP BY T1.STK_CD
        ORDER BY VOL_AMT DESC
        LIMIT 20                                               ################## 위의 쿼리를 인라인으로 처
        ) T3 # [SQL 5.12.2-2]를 인라인 뷰로 처리                # 이 쿼리의 결과를 T3라는 별칭의 테이블로 정의
        
        # 동일한 테이블에서 특정 날짜의 값을 따로 가져와야 하므로,
        # 아래 쿼리에서 HISTORY_DT 테이블의 별칭을 T4, T5로 따로 지정한다. (240517)
        INNER JOIN HISTORY_DT T4
           ON (T4.STK_CD = T3.STK_CD
               AND T4.DT = STR_TO_DATE('20190502','%Y%m%d'))    
               
        INNER JOIN HISTORY_DT T5      # 굳이 동일한 테이블을 재사용하면서 테이블 별칭을 새로 지정하는 이유는? (240516)
           ON (T5.STK_CD = T3.STK_CD
               AND T5.DT = STR_TO_DATE('20191230','%Y%m%d'))

## 연결된 두 개의 INNER JOIN은 병렬의 결합이다. (240517) - 그림을 그려서 생각할 것
               
ORDER BY T3.VOL_AMT DESC;                                  
'''

import pymysql

con = pymysql.connect(
    user='root',
    passwd=PASS,
    host='127.0.0.1',
    db='db_sqlstk',
    charset='utf8'
)

mycursor = con.cursor()

query = QUERY

mycursor.execute(query)
data = mycursor.fetchall()
con.close()

# 쿼리 결과를 데이터프레임으로 출력
import pandas as pd

df = pd.DataFrame(data) 
df.columns = ['STK_CD', 'STK_NM', 'VOL_AMT', 'C_PRC_0502', 'C_PRC_1230']   # 5/1 : 근로자의 날
df

# 3월 한 달간 거래량이 많은 종목의 연말 종가 확인
```
