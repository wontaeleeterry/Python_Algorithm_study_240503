파이썬으로 산점도와 선형회귀선 그리기

```python
# 그래프 그리기 위한 라이브러리 가져오기

import seaborn as sns; sns.set()
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 엑셀 데이터(CSV파일)가져오기

tips = pd.read_csv('C:/Users/Windows10/Desktop/MJ/seaborn_data/tips.csv')

#가져온 데이터 확인하기 

tips.head()
```

두 변수의 관계를 산점도와 선형회귀선(신뢰구간)으로 나타내기

```python
sns.relplot(x="total_bill" , y ='tip', data = tips)
```

```python
ax = sns.lmplot(x="total_bill" , y ='tip', data = tips)
```

```python
# 68% 신뢰구간표현하기
#ci = 68
ax = sns.lmplot(x="total_bill" , y ='tip', data = tips, ci = 68)
```

```python
# 신뢰구간 없애기
#ci = 68
ax = sns.lmplot(x="total_bill" , y ='tip', data = tips, ci = None)
```

```python
# 신뢰구간 변경 및 없애기
# ci = 68 (여기에 원하는 신뢰구간 값을 넣으면 됨) 또는 ci = None 으로 나타내면 된다.
ax = sns.lmplot(x="total_bill" , y ='tip', data = tips, ci = None)

# x,y축 이름변경하기
plt.xlabel("total bill ($)")
plt.ylabel("tip ($)")

# 그리드라인, 배경바꾸기
sns.set_style(style ="darkgrid")

# 그래프 사이즈 조정하기
plt.figure(figsize = (10,8))

# 그래프 저장하기
plt.savefig("Regression_line", format ='png', dpi =150)

```

두 변수의 관계에 관한 통계량 분석하기
선형회귀선 값 나타내기

```python
# 통계량 분석위한 라이브러리 가져오기

from sklearn.linear_model import LinearRegression

# 데이터 행렬로 만들기

total_bill = np.array(tips.loc[:, 'total_bill']).reshape((-1,1))
tip = np.array(tips.loc[:, 'tip']).reshape((-1,1))

# function 사용하여 선형회귀선 값 나타내기

model = LinearRegression()
model.fit(total_bill, tip) 
r_sq = model.score(total_bill, tip)
print ('coefficient of determination is ', r_sq)

```

```python
# p-value 계산하기

import statsmodels.api as sm
mod = sm.OLS(total_bill, tip)
fii = mod.fit()
p_value = fii.summary2().tables[1]['P>|t|']

print ('p-value for the regression is  ',  p_value)
print ('p-value for the regression is %0.9f ' % p_value)
```
