```python
# 등락률 계산하기

import pymysql

con = pymysql.connect(
    user='root',
    passwd=PASS,  # 윈도우 노트북 확인
    host='127.0.0.1',
    db='db_sqlstk',        # 윈도우 노트북 확인
    charset='utf8'
)


mycursor = con.cursor()

query = '''
SELECT  T1.STK_CD ,T1.STK_NM
        ,T2.DT ,T2.C_PRC
        ,T3.DT ,T3.C_PRC
        ,ROUND((T2.C_PRC - T3.C_PRC) / T3.C_PRC * 100,2) CHG_RT   # 계산 결과가 CHG_RT 컬럼 값이 된다.
FROM    STOCK T1
        INNER JOIN HISTORY_DT T2
          ON (T2.STK_CD = T1.STK_CD)
        INNER JOIN HISTORY_DT T3
          ON (T3.STK_CD = T2.STK_CD)
#WHERE   T1.STK_NM LIKE '삼성%'                     # 삼성 계열사
WHERE   T1.STK_NM IN ('삼성전자','카카오','LG화학')     # 특정 종목 여러개 지정
AND     T2.DT = STR_TO_DATE('20201230','%Y%m%d')
AND     T3.DT = STR_TO_DATE('20200102','%Y%m%d');
'''
mycursor.execute(query)
data = mycursor.fetchall()
con.close()

# 쿼리 결과를 데이터프레임으로 출력
import pandas as pd

df = pd.DataFrame(data) 
df.columns = ['종목코드', '종목명', '일자(시작)', '종가', '일자(종료)', '종가', '수익률(%)']
df

#display(data)
```

```python
# 하나의 SQL에 인라인 뷰를 여러 개 사용할 수 있는 것처럼 WITH절도 여러 개 사용 가능
# RES1에서는 2019년 2월에 등락률이 가장 큰 종목 열 개를 찾아내고, 
# RES2에서는 RES1에서 찾은 결과 중에 2월 28일 가격이 가장 작은 종목 세 개를 찾아내고 있다.

import pymysql

con = pymysql.connect(
    user='root',
    passwd=PASS,
    host='127.0.0.1',
    db='db_sqlstk',
    charset='utf8'
)

mycursor = con.cursor()

query = '''
WITH RES1 AS (
    SELECT  T1.STK_CD ,T1.STK_NM
            ,T2.DT ,T2.C_PRC 0228_가격
            ,T3.C_PRC 0201_가격       
            ,ROUND((T2.C_PRC - T3.C_PRC) / T3.C_PRC * 100,2) CHG_RT
    FROM    STOCK T1
            INNER JOIN HISTORY_DT T2 ON (T2.STK_CD = T1.STK_CD)
            INNER JOIN HISTORY_DT T3 ON (T3.STK_CD = T1.STK_CD)
    WHERE   T2.DT = STR_TO_DATE('20190228','%Y%m%d')
    AND     T3.DT = STR_TO_DATE('20190201','%Y%m%d')
    ORDER BY CHG_RT DESC
    LIMIT 10)
,RES2 AS(
    SELECT  T4.*
    FROM    RES1 T4
    ORDER BY T4.0228_가격 ASC
    LIMIT 3)
SELECT  T5.*
FROM    RES2 T5;
'''
mycursor.execute(query)
data = mycursor.fetchall()
con.close()

# 쿼리 결과를 데이터프레임으로 출력
import pandas as pd

df = pd.DataFrame(data) 
df.columns = ['종목코드', '종목명', '일자', '종가(2/28)', '종가(2/01)', '수익률(%)']
df

#display(data)

# 가격이 가장 작은 종목 3개를 찾아 2월 가격 등락 확인
```

```python
# 특정 종목의 거래량(VOL) 가장 작았던 데이터 3 건을 조회

import pymysql

con = pymysql.connect(
    user='root',
    passwd=PASS,
    host='127.0.0.1',
    db='db_sqlstk',
    charset='utf8'
)

mycursor = con.cursor()

query = '''
SELECT  T1.STK_NM ,T2.DT ,T2.VOL, T2.C_PRC   # STOCK 테이블에서 종목명, 날짜, 거래량 가져오기
FROM    STOCK T1                   # STOCK 테이블의 별칭을 T1으로 설정
        INNER JOIN HISTORY_DT T2       # 종목 테이블(WHERE에서 선정)과 그 종목의 HISTORY를 매칭하여 교집합 추출 (240514)
           ON (T2.STK_CD = T1.STK_CD)   
WHERE   T1.STK_NM = '삼성전자'      # 특정 조건 선정
ORDER BY T2.VOL ASC                # 오름 차순으로 정
LIMIT 10;                           # 거래량(VOL)을 기준으로 3건만 조회
'''
mycursor.execute(query)
data = mycursor.fetchall()
con.close()

# 쿼리 결과를 데이터프레임으로 출력
import pandas as pd

df = pd.DataFrame(data) 
df.columns = ['종목명', '일자', '거래량', '종가']
df

# 확률적으로 특정 기간에 거래량이 줄어드는지 여부 확인 가능(240514)
# 거래량과 종가는 상관관계가 있는가?
# 삼성전자의 종가와 KOSPI 상관관계?

# 여름(7월)에 거래량이 급감
```
