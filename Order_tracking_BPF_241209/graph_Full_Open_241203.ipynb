```python
import pandas as pd
data = pd.read_csv('test - Base_Ful - Mic_1m - Global - Tone-Tacho   9 O dB A  T 0.25  WV .csv', sep=",")
data_2 = pd.read_csv('test - Clear_Ful -  - Mic_1m - Global - Tone-Tacho   9 O dB A  T 0.25  WV .csv', sep=",")
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

plt.yticks(np.arange(40, 106, 5))
plt.xticks(np.arange(250*60, 1400*60, 5000))

# 표기 반대로 되어 있음 (241203)
# 1일차 - 클리어런스 증대 샘플 -> Base로 표기 -> 변경할 것
# 2일차 - 베이스 샘플 - Clear 로 표기 -> 변경할 것
plt.plot(data['l_Open_1 - 1 (rev/s)']*60, data['l_Open_1 - 1 (dB)'], alpha = 0.3, label='Clearance_Trial_#1')
plt.plot(data['l_Open_2 - 2 (rev/s)']*60, data['l_Open_2 - 2 (dB)'], alpha = 0.3, label='Clearance_Trial_#2')
plt.plot(data['l_Open_3 - 3 (rev/s)']*60, data['l_Open_3 - 3 (dB)'], alpha = 0.3, label='Clearance_Trial_#3')
plt.plot(data.iloc[:,-2]*60, data.iloc[:,-1], label='Clearance_Average')

plt.plot(data_2['4 (rev/s)']*60, data_2['4 (dB)'], alpha = 0.3, label='Base_Trial_#1')
plt.plot(data_2['5 (rev/s)']*60, data_2['5 (dB)'], alpha = 0.3, label='Base_Trial_#2')
plt.plot(data_2['6 (rev/s)']*60, data_2['6 (dB)'], alpha = 0.3, label='Base_Trial_#3')
plt.plot(data_2.iloc[:,-2]*60, data_2.iloc[:,-1], label='Base_Average', color='b')


plt.title('Mic. near TC @1m Away : Vane Full Open')
plt.xlabel('TC Speed (RPM)')
plt.ylabel('SPL of TBPF - 9th Order (dB(A))')
plt.xlim([250*60, 77000])      # X축의 범위: [xmin, xmax]
plt.ylim([45, 105])              # Y축의 범위: [ymin, ymax]
plt.grid(True, linestyle='--')
plt.legend(loc='lower right')
plt.show()
```
