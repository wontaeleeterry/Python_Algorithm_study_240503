```python
import pandas as pd

data = pd.read_csv('data_240618.csv', sep = ',', na_values = ["''"])
```

```python
data
```

```python
x = data.level_1
y = data.level_2
Rate = data.Pass_rate
Samples = data.Sample_No
```

```python
#### 맵 색상이 순서대로 적층되도록 for 문으로 작성

import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize = (10, 10))

# fo문으로 샘플 정보를 순차적으로 꺼내와서 적층하기

plt.fill_between([0, 3.5], [0, 3.5], alpha=1, color='green')
plt.fill_betweenx([0, 3.5], [0, 3.5], alpha=1, color='green')

for i in range(len(data)):
    if Rate[i] >= 55:
        plt.fill_between([0, x[i]], [0, y[i]], alpha=1, color='green')
        plt.fill_betweenx([0, y[i]], [0, x[i]], alpha=1, color='green')

    if Rate[i] < 55:
        plt.fill_between([0, x[i]], [0, y[i]], alpha=1, color='yellow')
        plt.fill_betweenx([0, y[i]], [0, x[i]], alpha=1, color='yellow')

    if Rate[i] > 45 and Rate[i] < 50:
        plt.fill_between([0, x[i]], [0, y[i]], alpha=1, color='gold')
        plt.fill_betweenx([0, y[i]], [0, x[i]], alpha=1, color='gold')

    if Rate[i] <= 45:
        plt.fill_between([0, x[i]], [0, y[i]], alpha=1, color='red')
        plt.fill_betweenx([0, y[i]], [0, x[i]], alpha=1, color='red')

# 컬러 박스의 플롯이 완료되면,
# 모든 샘플에 대한 Scattering 진행

CS = plt.scatter(x, y, color='black')

for i, txt in enumerate(Samples):
    ax.annotate(txt, (x[i]+0.03, y[i]), color='black', fontsize = 14)

plt.title('Graphical representation for Samples', fontsize = 16)
plt.xlabel('Low Freq. band range', fontsize = 14)
plt.ylabel('High Freq. band range', fontsize = 14)

# Rate annotation
ax.annotate('Rate', (0.05, 2.8), color = 'black', fontsize = 18)
ax.annotate('> 55%', (0.05, 2.6), color = 'black', fontsize = 16)
ax.annotate('50-55%', (0.05, 2.4), color = 'black', fontsize = 16)
ax.annotate('45-50%', (0.05, 2.1), color = 'black', fontsize = 16)
ax.annotate('< 45%', (0.05, 1.8), color = 'black', fontsize = 16)


plt.xlim(0, 3.5)
plt.ylim(0, 3.5)

plt.show()
```
